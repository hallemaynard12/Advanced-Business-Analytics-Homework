# 1**.	Compute the arithmetic mean of the electricity variable in the data.**

```r
# Installing Packages
#install.packages(c('vtable', 'readxl'))
library('vtable')
library('readxl')

# Reading the data in the Excel File
mkt_data <- read.csv('/content/sample_data/Electric_Production.csv')

elec_avg <- sum(mkt_data$Electricity / length(mkt_data$Electricity))

cat("The Arithmetic Mean is:", elec_avg)
```

# **2.	Compute the midrange of the electricity variable**

        a. Sometimes, midrange is not a good idea, why?
          Midrange ignores all intermediate values between the maximum and minimum values and is easily influenced by extreme values.
          This means that it is centralized around your outliars and other values that can drasticly change the data.
          This makes the value it produces can inaccurately represent the data.





```r
# Get Max and Min values of the Electricity variable
max_elec <- max(mkt_data$Electricity)
min_elec <- min(mkt_data$Electricity)

# Calculating the Midrange
mid_range <- 0.5*(min_elec + max_elec)
cat("The Midrange is:", mid_range)
```

# **3.	Find the geometric mean of electricity variable**

```r
# Get hte products of all numbers
elec_prod <- prod(mkt_data$Electricity)

# Calculate the nth root of the product
geoMean <- (elec_prod)^(1/length(mkt_data$Electricity))
cat("The Geometric Mean is:", geoMean)
```

# **4.	Find the harmonic mean of electricity variable**

```r
# Compute harmonic mean
harm_mean <- length(mkt_data$Electricity) / sum(1/mkt_data$Electricity)
cat("The Harmonic Mean is:", harm_mean)
```

# **5.	Picking any seed you want, generate a uniform distribution between 0.1 and 1.8. Use this as your weights. Find the weighted average of electricity variable**

```r
# Set Seed
set.seed(56)

# Add Weights
weights <- runif(length(mkt_data$Electricity), 0.1,1.8)

# Calculating the Weighted Average
weighted_avg <- sum(weights*mkt_data$Electricity)/sum(weights)
cat("The weighted average is:", weighted_avg)
```

# **6.	Compare all the mean values. Which measurement is always smallest? Which one is the biggest?**
  
          By looking at all the values and comparing them to other examples in
          class it is aparent that the Harmonic Mean is always the smallest and
          the Weighted average is always the largest.



```r
cat("The Arithmetic Mean is:", elec_avg)
cat("\nThe Midrange is:", mid_range)
cat("\nThe Geometric Mean is:", geoMean)
cat("\nThe Harmonic Mean is:", harm_mean)
cat("\nThe weighted average is:", weighted_avg)
```

# **7.	Compute the median and trimmed mean of the electricity variable. Trim 1%**

```r
# Compute the Median and Mean
elec_median <- median(mkt_data$Electricity)
elec_mean <- mean(mkt_data$Electricity)

# Trim the mean by 1%
trim_mean <- mean(mkt_data$Electricity, trim = 0.01)

# Print
cat("Mean:", elec_mean)
cat("\nMedian:", elec_median)
cat("\nTrimmed Mean:", trim_mean)
```

# **8.	Compute a winsorized mean using a 2.5% for th bottom percentile and 97.5% for the upper percentile**

```r
# Install a Package for Winsorization
#install.packages("statar")
library("statar")

# Preform winsorization
num_winsor <- winsorize(mkt_data$Electricity, probs = c(0.025, 0.975), verbose = FALSE)
mean_winsor <- mean(num_winsor)

# Print
cat("Winsorized Mean:", mean_winsor)
```
