## **1. Build a classification model to predict whether the tumor is malignant (“M”) or benign (“B”)**

> a. Use all the classification models we have discussed in class.

> b. Use a threshold value of 0.5 to decide the class

```r
# Must begin by importing and reading the data
cancer_data <- read.csv('/content/sample_data/breast cancer.csv')

# Install Packages
#install.packages('MASS')
library('MASS')

#install.packages('e1071')
library('e1071')

# Convert dependent variable, diagnosis, to be a factor
cancer_data$diagnosis <- as.factor(cancer_data$diagnosis)

# Naieve Bays, logistic regerssion and linear discriminat analysis model
 nb_model <- naiveBayes(diagnosis ~., data=cancer_data, type='class')
 lr_model <- glm(diagnosis ~., data=cancer_data, family='binomial')
 lda_model <- lda(diagnosis ~., data=cancer_data)
```

```r
# Getting predictions from all the classification models
# Predict NB Model
nb_pred <- predict(nb_model, newdata=cancer_data, type='class')

# Predict LR Model
lr_pred <- predict(lr_model, type='response')

# Predict LDA Model
lda_pred <- predict(lda_model, newdata=cancer_data)
```

```r
# Class labels for LR model
lr_class <- ifelse(lr_pred > 0.5, "M", "B")

# Class labels for LDA model
lda_prob <- as.data.frame(lda_pred$posterior)
lda_prob$class <- ifelse(lda_prob$B > 0.5, "M", "B")
```
